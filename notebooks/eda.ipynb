## Exploratory Data Analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# dataset
df = pd.read_csv("C:/Users/harsh/ensemble-learning-project/data/Ola_dataset.csv")


#  basic info
print(df.info())
df.head()
```

```python
original=df.copy()
```

```python
# Check for missing values
print(df.isnull().sum())
```

```python
# basic statistical details of numeric columns
df.describe()
```

```python
# histograms for all numeric columns
df.hist(figsize=(12, 8), bins=20)
plt.show()
```

```python
# correlation heatmap to see the relationships between numeric features
plt.figure(figsize=(12, 8))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm", fmt=".2f")
plt.show()
```

```python
# count of unique values in categorical columns
for col in df.select_dtypes(include='object').columns:
    plt.figure(figsize=(10, 5))
    sns.countplot(x=col, data=df)
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Boxplot to check for outliers in numerical columns
plt.figure(figsize=(10, 6))
sns.boxplot(data=df)
plt.xticks(rotation=270)
plt.show()
```

```python
df.columns
```

```python
# histograms for numerical columns
numerical_cols = ['Age', 'Total Business Value', 'Income', 'Quarterly Rating', 'Income_raised']

df[numerical_cols].hist(figsize=(12, 8), bins=20)
plt.tight_layout()
plt.show()
```

```python
# Scatter plot to show relationship between Age and Income
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Age', y='Income', data=df, hue='Gender', palette='coolwarm')
plt.title('Age vs Income by Gender')
plt.show()
```

```python
# Boxplot for target variable vs numerical features
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='Income', data=df)
plt.title('Income Distribution by Target')
plt.show()
```

```python
# Countplot for categorical columns
categorical_cols = ['Gender', 'Education_Level', 'City', 'Joining Designation']

for col in categorical_cols:
    plt.figure(figsize=(10, 6))
    sns.countplot(x=col, data=df)
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45)
    plt.show()
```

```python
# Bar plot for Income vs Target
plt.figure(figsize=(10, 6))
sns.barplot(x='target', y='Income', data=df)
plt.title('Average Income by Target')
plt.show()
```

```python
# for categorical features: Education Level vs Gender
contingency_table = pd.crosstab(df['Education_Level'], df['Gender'])
plt.figure(figsize=(10, 6))
sns.heatmap(contingency_table, annot=True, cmap='Blues', fmt='d')
plt.title('Education Level vs Gender')
plt.show()
```

```python
# Boxplot for target vs quarterly rating
plt.figure(figsize=(10, 6))
sns.boxplot(x='target', y='Quarterly Rating', data=df)
plt.title('Quarterly Rating by Target')
plt.show()
```

```python
# Line plot for Total Business Value by Joining Year
plt.figure(figsize=(12, 6))
sns.lineplot(x='joining_year', y='Total Business Value', data=df)
plt.title('Total Business Value over Joining Year')
plt.show()
```

```python
# Heatmap for quaterly_rating_raise with other features
plt.figure(figsize=(12, 8))
sns.heatmap(df[['Income', 'Quarterly Rating', 'Income_raised', 'quaterly_rating_raise']].corr(), annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap for Quarterly Rating Raise and Other Features')
plt.show()
```
